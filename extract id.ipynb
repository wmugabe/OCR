```python
import cv2
import pytesseract
from PIL import Image
import re
import os

# Set the Tesseract OCR path (adjust this path to your installation)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# Specify the full path to the image
image_path = r"C:\Users\danie\Downloads\id.png"

# Check if the file exists
if not os.path.exists(image_path):
    raise FileNotFoundError(f"File not found: {image_path}")

# Load the image
image = cv2.imread(image_path)

# Verify the image is loaded
if image is None:
    raise ValueError("Failed to load image. Check the file path or file format.")

# Preprocess the image
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to binarize the image
thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

# Save the preprocessed image (optional, for debugging)
preprocessed_path = r"C:\Users\danie\Downloads\preprocessed_image.jpg"
cv2.imwrite(preprocessed_path, thresh)
print(f"Preprocessed image saved at: {preprocessed_path}")

# Extract text from the preprocessed image
extracted_text = pytesseract.image_to_string(thresh)

# Display the extracted text
print("Extracted Text:")
print(extracted_text)

# Optional: Extract specific fields using regular expressions
# Example: Find an ID number pattern (e.g., 6-10 consecutive digits)
id_number = re.search(r'\b\d{6,10}\b', extracted_text)
if id_number:
    print(f"ID Number Found: {id_number.group()}")
else:
    print("No ID Number Found.")

# Example: Extract a name if it follows a specific pattern (e.g., "Name: John Doe")
name_match = re.search(r'Name:\s*([A-Za-z ]+)', extracted_text)
if name_match:
    print(f"Name Found: {name_match.group(1)}")
else:
    print("No Name Found.")

# Optional: Crop specific regions of the card and extract text from them
# (Useful for cards with structured layouts)
# Example: Crop and OCR a region (adjust coordinates as needed)
roi = image[100:300, 200:500]  # Adjust based on your card's layout
roi_text = pytesseract.image_to_string(roi)
print("Text from ROI (Region of Interest):")
print(roi_text)
```
